```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
      .appName("jdbc-test")
      .config("spark.jars.packages", "org.postgresql:postgresql:42.7.4")  # подтянет jar
      .getOrCreate()
)
```

```python
pg_options = {
    "url": "jdbc:postgresql://postgres:5432/dwh",   # ВАЖНО: хост = имя контейнера, не localhost
    "dbtable": "stg.orders",
    "user": "app",
    "password": "app",
    "driver": "org.postgresql.Driver"
}
```

```python
df = spark.read.format("jdbc").options(**pg_options).load()
df.printSchema()
df.show(5, truncate=False)
```

```python

# 1) Кол-во заказов по статусу
df.groupBy("order_status").count().orderBy("count", ascending=False).show()
```

```python

# 2) Среднее время доставки (от purchase до delivered)
from pyspark.sql.functions import col, unix_timestamp
```

```python

orders_delivered = df.where(col("order_delivered_customer_date").isNotNull())
agg = orders_delivered.select(
    (unix_timestamp("order_delivered_customer_date") - unix_timestamp("order_purchase_ts")).alias("sec")
).groupBy().avg("sec")

agg.selectExpr("round(avg(sec)/3600, 2) as avg_hours_to_deliver").show()
```
